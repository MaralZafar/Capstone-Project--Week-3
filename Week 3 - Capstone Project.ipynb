## 2. scrape the wikipedia page to transfer the data

```python
#scrape the wikipedia page to transfer the data
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(source,'lxml')
print(soup.title)
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

```python
#The top 5 rows
dfs = pd.read_html(tab)
df=dfs[0]
df.head()
```

```python
# Number of rows and coulumn for main df
df.shape
```

## 3.Cleaned data without not assigned neighborhood and sorted with index

```python
#the cleaned data without not assigned neighborhood and sorted with index 
df1 = df[df.Borough != 'Not assigned'].reset_index(drop=True)
df1
```

```python
# Number of rows and coulumn for df1
df1.shape
```

```python
lat_lon = pd.read_csv('https://cocl.us/Geospatial_data')
lat_lon.head()
```

```python
lat_lon.shape
```

```python
df2 = df1.groupby(['Postal Code','Borough'], sort=False).agg(', '.join)
df2.reset_index(inplace=True)
df2
```

## 4. Merging two tables to have lat&long beside Sorted postal code

```python
#Merged 2 tables to have lat&long beside Sorted postal code
mask = df2['Neighborhood'] == "Not assigned"
df2.loc[mask, 'Neighborhood'] = df2.loc[mask, 'Borough']
df2['Neighborhood'] = np.where(df2['Neighborhood'] == 'Not assigned',df2['Borough'], df2['Neighborhood'])
df2['Latitude'] = lat_lon ['Latitude']
df2['Longitude'] = lat_lon ['Longitude']
df2
```

```python
#Merged tables dimension
df2.shape
```

## 5. Merged tables cluster the neighborhood in Toronto

```python
#Merged tables cluster the neighborhood in Toronto
df3 = df2[df2['Borough'].str.contains('Toronto',regex=False)]
df3
```

```python
#Merged tables dimension around the Torontos neighborhood
df3.shape
```

## 6. Plot map

```python
map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,Neighborhood in zip(df3['Latitude'],df3['Longitude'],df3['Borough'],df3['Neighborhood']):
    label = '{}, {}'.format(Neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
    
map_toronto
```

## 7. Clustering

```python
k=5
toronto_clustering = df3.drop(['Postal Code','Borough','Neighborhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_[0:10]
```

```python
# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]
```

```python
# create map
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, Neighborhood, cluster in zip(df3['Latitude'], df3['Longitude'], df3['Neighborhood'], df3['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
```
